## Adicionar colunas com os dados tratados

As variáveis importantes serão: número do órgão, nome do órgão, número da unidade gestora, nome da unidade gestora, CNPJ, nome do fornecedor e número do contrato sem barra.

```python
# Separando as colunas para termos as informações do órgão
df_terc['nr_orgao'] = df_terc['Órgão'].str.split(' - ').str[0].str.strip()
df_terc['nome_orgao'] = df_terc['Órgão'].str.split(' - ').str[1].str.strip()
```

```python
# Separando as colunas para termos as informações da unidade gestora
df_terc['nr_ug'] = df_terc['Unidade Gestora'].str.split(' - ').str[0].str.strip()
df_terc['nome_ug'] = df_terc['Unidade Gestora'].str.split(' - ').str[1].str.strip()
```

```python
# Separando as colunas para termos as informações dos fornecedores
df_terc['cnpj'] = df_terc['Fornecedor'].str.split(' - ').str[0].str.strip()
df_terc['fornecedor'] = df_terc['Fornecedor'].str.split(' - ').str[1].str.strip()
```

```python
# Retirar a barra do número de contrato
df_terc['nr_contrato'] = df_terc['Número Contrato'].str.replace('/', '', regex=False)
```

```python
# Excluindo as colunas originais
df_terc = df_terc.drop(columns=['Órgão', 'Unidade Gestora', 'Fornecedor', 'Número Contrato'])
```

## Calcular o quantitativo de terceirizados por contrato, empresa e órgão

```python
df_contratos = (
    df_terc
    .groupby(['nr_contrato', 'cnpj', 'nr_orgao', 'nr_ug'], as_index=False)
    .size()
    .rename(columns={'size': 'quant_terceirizados'})
)
```
